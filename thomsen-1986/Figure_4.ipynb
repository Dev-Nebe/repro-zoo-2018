The data file from Table 1 is in this directory. Matt digitized it a while ago and put it [in Figshare](https://figshare.com/articles/Thomsen_1986_dataset/1304337). 

```python
import pandas as pd

df = pd.read_csv('Thomsen_1986_Table1.csv')
```

```python
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(df.epsilon, df.delta)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.show()
```
